```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce

# -----------------------------
# 1️⃣ Helper function: create binary indicator per repo for a feature
# -----------------------------
def make_feature_df(dfs, feature_name):
    df_list = []
    for df in dfs:
        tmp = df[['full_name']].copy()
        tmp[feature_name] = 1
        df_list.append(tmp)
    if df_list:
        return pd.concat(df_list, ignore_index=True).drop_duplicates()
    return pd.DataFrame(columns=['full_name', feature_name])

# -----------------------------
# 2️⃣ Define features
# -----------------------------

# Authentication split into peer and request
peer_auth_dfs = [df_peer_auth_v1, df_peer_auth_v1beta1, df_peer_auth_v1alpha1]
request_auth_dfs = [df_req_auth_v1, df_req_auth_v1beta1, df_req_auth_v1alpha1]

df_peer_auth = make_feature_df(peer_auth_dfs, 'Peer')
df_request_auth = make_feature_df(request_auth_dfs, 'Request')

# Authorization
authz_dfs_dict = {
    'HTTP': [df_http_authz_v1, df_http_authz_v1beta1],
    'TCP': [df_tcp_authz_v1, df_tcp_authz_v1beta1],
    'JWT': [df_jwt_authz_v1, df_jwt_authz_v1beta1],
    'Provider': [df_provider_authz_v1, df_provider_authz_v1beta1],
    'Ingress': [df_ingress_authz_ip_v1, df_ingress_authz_ip_v1beta1,
                df_ingress_authz_remote_ip_v1, df_ingress_authz_remote_ip_v1beta1],
    'Any': [df_any_authz_v1alpha1]
}

df_authz_list = [make_feature_df(dfs, feature) for feature, dfs in authz_dfs_dict.items()]

# mTLS modes
mtls_dfs_dict = {
    'Strict': [df_strict_v1, df_strict_v1beta1, df_strict_v1alpha1],
    'Permissive': [df_permissive_v1, df_permissive_v1beta1, df_permissive_v1alpha1],
    'Disable': [df_disable_v1, df_disable_v1beta1]
}

df_mtls_list = [make_feature_df(dfs, mode) for mode, dfs in mtls_dfs_dict.items()]

# -----------------------------
# 3️⃣ Merge all feature indicators by repository
# -----------------------------
dfs_to_merge = [df_peer_auth, df_request_auth] + df_authz_list + df_mtls_list
df_features = reduce(lambda left, right: pd.merge(left, right, on='full_name', how='outer'), dfs_to_merge)
df_features = df_features.fillna(0)

# -----------------------------
# 4️⃣ Compute co-occurrence matrix
# -----------------------------
feature_cols = df_features.columns.drop('full_name')
co_occurrence = df_features[feature_cols].T.dot(df_features[feature_cols])

# Optional: normalize to percentage of repos
co_occurrence_pct = co_occurrence / len(df_features) * 100

# -----------------------------
# 5️⃣ Plot heatmap
# -----------------------------
plt.figure(figsize=(12,10))
sns.heatmap(co_occurrence_pct, annot=True, fmt=".1f", cmap="Blues")
plt.title("Co-occurrence Heatmap: mTLS, Authentication & Authorization", fontproperties=prop)
plt.xticks(rotation=45, ha='right', fontproperties=prop)
plt.yticks(rotation=0, fontproperties=prop)
plt.tight_layout()
plt.savefig("./figures/total_heatmap.svg", dpi=300, bbox_inches="tight", pad_inches=0)

plt.show()
```
