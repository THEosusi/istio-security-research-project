```python
import pandas as pd

file_path = "../data/repositoryinfo/http_authz_v1_repos.parquet"

df_http_authz_v1 = pd.read_parquet(file_path)
df_http_authz_v1

file_path = "../data/repositoryinfo/http_authz_v1beta1_repos.parquet"
df_http_authz_v1beta1 = pd.read_parquet(file_path)
df_http_authz_v1beta1

file_path = "../data/repositoryinfo/tcp_authz_v1_repos.parquet"
df_tcp_authz_v1 = pd.read_parquet(file_path)
df_tcp_authz_v1

file_path = "../data/repositoryinfo/tcp_authz_v1beta1_repos.parquet"
df_tcp_authz_v1beta1 = pd.read_parquet(file_path)
df_tcp_authz_v1beta1

file_path = "../data/repositoryinfo/jwt_authz_v1_repos.parquet"
df_jwt_authz_v1 = pd.read_parquet(file_path)
df_jwt_authz_v1

file_path = "../data/repositoryinfo/jwt_authz_v1beta1_repos.parquet"
df_jwt_authz_v1beta1 = pd.read_parquet(file_path)
df_jwt_authz_v1beta1

file_path = "../data/repositoryinfo/provider_authz_v1_repos.parquet"
df_provider_authz_v1 = pd.read_parquet(file_path)
df_provider_authz_v1

file_path = "../data/repositoryinfo/provider_authz_v1beta1_repos.parquet"
df_provider_authz_v1beta1 = pd.read_parquet(file_path)
df_provider_authz_v1beta1

file_path = "../data/repositoryinfo/ingress_authz_ip_v1_repos.parquet"
df_ingress_authz_ip_v1 = pd.read_parquet(file_path)
df_ingress_authz_ip_v1

file_path = "../data/repositoryinfo/ingress_authz_ip_v1beta1_repos.parquet"
df_ingress_authz_ip_v1beta1 = pd.read_parquet(file_path)
df_ingress_authz_ip_v1beta1

file_path= "../data/repositoryinfo/ingress_authz_remote_ip_v1_repos.parquet"
df_ingress_authz_remote_ip_v1 = pd.read_parquet(file_path)
df_ingress_authz_remote_ip_v1

file_path = "../data/repositoryinfo/ingress_authz_remote_ip_v1beta1_repos.parquet"
df_ingress_authz_remote_ip_v1beta1 = pd.read_parquet(file_path)
df_ingress_authz_remote_ip_v1beta1


file_path = "../data/repositoryinfo/any_authz_v1alpha1_repos.parquet"
df_any_authz_v1alpha1 = pd.read_parquet(file_path)
df_any_authz_v1alpha1
```

```python
df_all = pd.concat([df_http_authz_v1, df_http_authz_v1beta1, df_tcp_authz_v1,df_tcp_authz_v1beta1,df_jwt_authz_v1, df_jwt_authz_v1beta1, df_provider_authz_v1,df_provider_authz_v1beta1,df_ingress_authz_ip_v1,df_ingress_authz_ip_v1beta1,df_ingress_authz_remote_ip_v1,df_ingress_authz_remote_ip_v1beta1,df_ingress_authz_remote_ip_v1alpha1], ignore_index=True)
df_v1 = pd.concat([df_http_authz_v1, df_tcp_authz_v1, df_jwt_authz_v1, df_provider_authz_v1, df_ingress_authz_ip_v1, df_ingress_authz_remote_ip_v1], ignore_index=True)
df_v1beta1 = pd.concat([df_http_authz_v1beta1, df_tcp_authz_v1beta1, df_jwt_authz_v1beta1, df_provider_authz_v1beta1, df_ingress_authz_ip_v1beta1, df_ingress_authz_remote_ip_v1beta1], ignore_index=True)
df_v1alpha1 = pd.concat([df_any_authz_v1alpha1], ignore_index=True)
df_http_authz=pd.concat([df_http_authz_v1,df_http_authz_v1beta1], ignore_index=True)
df_tcp_authz=pd.concat([df_tcp_authz_v1,df_tcp_authz_v1beta1], ignore_index=True)
df_jwt_authz=pd.concat([df_jwt_authz_v1,df_jwt_authz_v1beta1], ignore_index=True)
df_provider_authz=pd.concat([df_provider_authz_v1,df_provider_authz_v1beta1], ignore_index=True)
df_ingress_authz=pd.concat([df_ingress_authz_ip_v1,df_ingress_authz_ip_v1beta1,df_ingress_authz_remote_ip_v1,df_ingress_authz_remote_ip_v1beta1], ignore_index=True)


df_all_unique = df_all.drop_duplicates(subset="full_name")
df_v1_unique = df_v1.drop_duplicates(subset="full_name")
df_v1beta1_unique = df_v1beta1.drop_duplicates(subset="full_name")
df_v1alpha1_unique = df_v1alpha1.drop_duplicates(subset="full_name")
df_http_authz_unique = df_http_authz.drop_duplicates(subset="full_name")
df_tcp_authz_unique = df_tcp_authz.drop_duplicates(subset="full_name")
df_jwt_authz_unique = df_jwt_authz.drop_duplicates(subset="full_name")
df_provider_authz_unique = df_provider_authz.drop_duplicates(subset="full_name")
df_ingress_authz_unique = df_ingress_authz.drop_duplicates(subset="full_name")

print("All:", len(df_all_unique))
print("v1:", len(df_v1_unique)-len(df_v1beta1_unique))
print("v1beta1:", len(df_v1beta1_unique))
print("v1alpha1:", len(df_v1alpha1_unique))
print("http_authz:", len(df_http_authz_unique))
print("tcp_authz:", len(df_tcp_authz_unique))
print("jwt_authz:", len(df_jwt_authz_unique))
print("provider_authz:", len(df_provider_authz_unique))
print("ingress_authz:", len(df_ingress_authz_unique))
```
