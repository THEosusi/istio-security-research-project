```python
# --- LaTeX table with separate Unique / Starred columns ---
versions = ["v1", "v1beta1", "v1alpha1"]

# Short labels for rows
kinds = {
    "http_authz": "HTTP",
    "tcp_authz": "TCP",
    "jwt_authz": "JWT",
    "provider_authz": "Provider",
    "ingress_authz": "Ingress",
}

# Build sets of repo full_names for UNIQUE (per version)
version_sets_unique = {
    "v1": set(df_v1_unique["full_name"].unique()),
    "v1beta1": set(df_v1beta1_unique["full_name"].unique()),
    "v1alpha1": set(df_v1alpha1_unique["full_name"].unique()),
}

# Build sets of repo full_names for STARRED (per version)
version_sets_star = {
    "v1": set(df_v1_star["full_name"].unique()),
    "v1beta1": set(df_v1beta1_star["full_name"].unique()),
    "v1alpha1": set(df_v1alpha1_star["full_name"].unique()),
}

# Build sets for each AUTHZ kind (unique)
kind_sets_unique = {
    "http_authz": set(df_http_authz_unique["full_name"].unique()),
    "tcp_authz": set(df_tcp_authz_unique["full_name"].unique()),
    "jwt_authz": set(df_jwt_authz_unique["full_name"].unique()),
    "provider_authz": set(df_provider_authz_unique["full_name"].unique()),
    "ingress_authz": set(df_ingress_authz_unique["full_name"].unique()),
}

# Build sets for each AUTHZ kind (starred)
kind_sets_star = {
    "http_authz": set(df_http_authz_star["full_name"].unique()),
    "tcp_authz": set(df_tcp_authz_star["full_name"].unique()),
    "jwt_authz": set(df_jwt_authz_star["full_name"].unique()),
    "provider_authz": set(df_provider_authz_star["full_name"].unique()),
    "ingress_authz": set(df_ingress_authz_star["full_name"].unique()),
}

# Denominators per column (version totals)
totals_unique = {v: len(version_sets_unique[v]) for v in versions}
totals_star = {v: len(version_sets_star[v]) for v in versions}

# Build LaTeX table
latex_table = r"""\begin{table}[h!]
\centering
\begin{tabular}{lcccccc}
\hline
 & \multicolumn{2}{c}{v1} & \multicolumn{2}{c}{v1beta1} & \multicolumn{2}{c}{v1alpha1} \\
 & U & S & U & S & U & S \\
\hline
"""

for kind_key, row_label in kinds.items():
    row = row_label
    for v in versions:
        # Intersection counts
        unique_count = len(version_sets_unique[v] & kind_sets_unique[kind_key])
        star_count = len(version_sets_star[v] & kind_sets_star[kind_key])

        # Percentages relative to column totals
        tu = totals_unique[v]
        ts = totals_star[v]
        pct_unique = (unique_count / tu * 100) if tu > 0 else 0.0
        pct_star = (star_count / ts * 100) if ts > 0 else 0.0

        row += f" & {unique_count} ({pct_unique:.1f}\\%) & {star_count} ({pct_star:.1f}\\%)"
    row += r" \\"
    latex_table += row + "\n"

latex_table += r"""\hline
\end{tabular}
\caption{Unique (U) and starred (S, $\geq$10 stars) repositories using authorization features across API versions. Percentages are relative to the total repositories in each version.}
\label{tab:authz_versions_star_split}
\end{table}"""

print(latex_table)
```
